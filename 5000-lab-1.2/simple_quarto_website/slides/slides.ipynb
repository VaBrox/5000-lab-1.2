import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")
print(df)

sns.set_theme(style = 'white', palette = "Pastel1")
g = sns.relplot(data = df, x = "bill_length_mm", y = "flipper_length_mm", hue = "species", col = "island")
FS = 18 
plt.setp(g._legend.get_texts(), fontsize=FS)
plt.setp(g._legend.get_title(), fontsize=FS)
plt.savefig('plot-1.pdf', bbox_inches='tight')
plt.show()
